```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('noshowappointments-kagglev2-may-2016.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.query("Age >= 0", inplace= True)
```

```python
df['ScheduledDay'] = pd.to_datetime(df.ScheduledDay)
df['AppointmentDay'] = pd.to_datetime(df.AppointmentDay)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
def column_bins(column, bins, labels):
    return pd.cut(column, bins, labels=labels, include_lowest=True)
```

```python
bins = [0, 18, 37, 55, 115]
labels = ["0-17", "18-36", "37-54", "55+"]
df["Age_Band"] = column_bins(df.Age, bins, labels)
```

```python
df.info()
```

```python
df.Age_Band.value_counts()
```

```python
df['Days_Difference'] = (df.AppointmentDay - df.ScheduledDay).astype('timedelta64[D]')
```

```python
df.describe()
```

```python
df.Days_Difference.value_counts()
```

```python
df[df.Days_Difference < 0]['Days_Difference'].unique()
```

```python
df[df.Days_Difference < -1]
```

```python
df.query('Days_Difference >= -1', inplace=True)
```

```python
df[df.Days_Difference < 0]
```

```python
df.Days_Difference.replace(-1.0, 0, inplace=True)
```

```python
df.describe()
```

```python
bins = [0, 3, 14, 178]
labels = ["0-2", "3-13", "14+"]
df["Days_Difference_Band"] = column_bins(df.Days_Difference, bins, labels)
```

```python
df.Days_Difference_Band.value_counts()
```

```python
df['Appointment_DOW'] = df.AppointmentDay.dt.dayofweek
df['Scheduled_DOW'] = df.ScheduledDay.dt.dayofweek
```

```python
df.head(10)
```

```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
def dayNameFromWeekday(weekday):
    days_ = []
    for day in weekday:
        days_.append(days[day])
    return days_
```

```python
df['Appointment_DOW'] = dayNameFromWeekday(df.Appointment_DOW)
df['Scheduled_DOW'] = dayNameFromWeekday(df.Scheduled_DOW)
```

```python
df.head(10)
```

```python
df['Appointment_DOW'] = pd.Categorical(df.Appointment_DOW, categories=days, ordered=True)
```

```python
df['Scheduled_DOW'] = pd.Categorical(df.Scheduled_DOW, categories=days, ordered=True)
```

```python
df.head(10)
```

```python
df.plot('Age','Days_Difference', kind='scatter');
plt.title('Relationship between Age and Days_Difference')
```
